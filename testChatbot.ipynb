```python
m = MeCab.Tagger()
resultSentences = []

for item in sentences:
    itemResult = []
    subjectWords = []
    verbWords = []
    
    # print(item)
    stritem = ''.join(item)
    # print(stritem)
    keywords = m.parse(stritem)
    # print(type(stritem))
    # print(keywords)
    # list1 = [1, 2, 3]
    # str1 = ''.join(str(e) for e in list1)
    
    for row in keywords.split("\n"):
        word = row.split("\t")[0]
        if word == "EOS":
            break
        else:
            pos = row.split("\t")[1].split(",")[0]
            pos2 = row.split("\t")[1].split(",")[1]
            ver = row.split("\t")[1].split(",")[6]
            if pos == "名詞":
                #if (pos2 != "固有名詞") and (pos2 != "接尾"):
                subjectWords.append(word)
            elif pos == "動詞":
                verbWords.append(ver)
    
    strSubjectWords = ' '.join(subjectWords)
    strVerbWords = ' '.join(verbWords)
    itemResult.append(stritem)
    itemResult.append(strSubjectWords)
    itemResult.append(strVerbWords)
    
    resultSentences.append(','.join(itemResult))

print(resultSentences)
```
